# Tipos de datos

En python tenemos los siguientes tipos de datos básicos: 

1. Números


2. Cadenas de caracteres (Strings)


3. Booleanos


4. None


## Números 

| Clase   |      Descripción      |  Ejemplo |
|----------|:-------------:|------:|
|integer| Número entero con precisión fija  | 45 |
| float |  Número decimal     |2.19   |
| complejo | Parte real y parte imaginaria |   (4.5 + 3j)  |

### Integer

Los *integer* o números enteros son aquellos que no tienen decimales, tanto positivos como negativos (además del cero). En Python se representan mediante el tipo `int`

````python
edad = 27
````

````python
print(edad, type(edad))
````

````python
edad = 0
````

````python
print(edad, type(edad))
````

````python
edad = -3
````

````python
print(edad, type(edad))
````

### Float o números reales

Los números reales son los que tienen decimales. En Python se expresan como tipo `float`

````python
precio = 33.9
````

````python
print(precio, type(precio))
````

````python
precio2 = -3.6
````

````python
print(precio2, type(precio))
````

````python
precio3 = .6
````

````python
print(precio3, type(precio))
````

### Complejo

Los números complejos son aquellos que tienen parte imaginaria. En Python se expresan como `complex`

````python
complejo = 2.1 + 7.8j
````

````python
print(complejo, type(complejo))
````

## Operadores de asignaciones

Imaginamos que tenemos una variable a la que le queremos sumar otro numero. Con los operadores anteriores lo podemos hacer


````python
r = 10
r = r + 5
print(r)
````

----
En Python tenemos los operadores de asignaciones que nos permiten realizar estas operaciones de una forma más sencilla, usando el operador `=`. 


>`+=` , `-=`, `*=`,  `/=`, `**=`, `//=`, `%=`: calcula la suma, resta,  multiplicación, división, exponente, división entera o módulo de la variable del lado izquierdo el valor del lado derecho

````python
r = 10
````

````python
r += 5
print(r)
````

````python
# ejemplo de cuando podemos usar este tipo de operador
count = 1
variable = 10
for _ in range(15):
    if count < variable:
        print("hola")
        count = count + 1 # es lo mismo que count += 1
        print(count)
    else:
        print("adios")
````

## Operadores relacionales

Se emplean para comparar y establecer la relación entre dos elementos. Devuelven un valor booleano (True o False) basado en la condición. 

`==`: evalua que los valores sean iguales para varios tipos de datos

````python
5 == 3
````

````python
if 5 == 3:
    print("error") # como es diferente, la condición no se cumple no sale nada en pantalla
else: # como la condición no se cumple pasamos al else y en este caso si se printea
    print("son diferentes")
````

````python
3 == 3 
````

````python
"avión" == 5
````

````python
"avión" == "avión"
````

````python
"avión" == "Avión" # ES CASE SENSITIVE!!
````

````python
True == False
````

````python
type("avión") == str
````

`!=`: evalua si los valores son distintos

````python
5 != 3
````

````python
"avión" != 3
````

````python
"avión" != "avión"
````

````python
True != False
````

`>`: evalua soi el valor del lado izquierdo es mayor que el valor del lado derecho

````python
5 > 3
````

````python
3 > 5
````

También podemos combinarlo con el operador =. `>=` que evalua si el valor del lado izquierdo es mayor o igual que el del lado derecho

````python
3 >= 5
````

````python
3 >= 3
````

`<`: evalua si el valor del lado izquierdo es menor que el lado derecho

````python
5 < 3
````

````python
3 < 5
````

De la misma que en el operador anterior tenemos el `<=` que evalua si el valor del lado izquierdo es menor o igual que el valor del lado derecho. 

````python
3 <= 3
````

````python
5 <= 3
````

## Otros operadores

### Operador de afiliciación 

Son operadores utilizados para chequear la pertenencia de un valor. Comprueba la pertenencia a una secuencia, como cadenas, listas o tuplas. 

````python
# el range por defecto empieza en 0
list_1 = list(range(10))
list_1
````

````python
# genera números empezando en el 1 hasta el 9
lst_ = list(range(1,10))
lst_
````

````python
lista = range(1,10)
````

````python
x = 5
y = 20
````

````python
x in lista
````

````python
y in lista
````

````python
x not in lista
````

````python
y not in lista
````

# Conversiones 

Las principales conversiones que podemos hacer en python son: 

- Convertir a *string*: str()


- Convertir a entero: int()


- Convertir a *float* (números decimales): float()


- Convertir a booleano: bool()

## Conversión de entero, decimal, y booleano a string 

````python
num1 = 23
num2 = 3.7
bool_ = True
````

````python
str(num1)
````

````python
str(num2)
````

````python
str(bool_)
````

## Conversión de entero y string a *float*

````python
num3 = 12
stri1 = "13.9"
````

````python
float(num3)
````

````python
float(stri1)
````

## Conversión de entero, decimal, y string a booleano

````python
num4 = 0
num5 = 10
num6 = 3.19
stri2 = ""
stri3 = "Hola"
````

````python
bool(num4)
````

````python
bool(num5)
````

````python
bool(num6)
````

````python
bool(stri2)
````

````python
bool(stri3)
````

📌 De estos ejemplos concluimos que los valores vacíos o nulos, cuando se les aplica la función bool(), se convierten en el False booleano. Por el contrario, cualquier otro valor equivale a True.
