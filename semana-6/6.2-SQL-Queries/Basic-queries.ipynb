# Queries básicas:

## Seleccionamos todas las columnas de una tabla 

```SQL
SELECT * -- con el asterisco indicamos que queremos todas las columnas
FROM nombre_tabla; -- de que tabla queremos que nos traiga la información
```

## Seleccionamos algunas columnas:

```sql
SELECT columna_1, columna_2 -- de que columnas queremos la información
FROM nombre_tabla; -- de que tabla queremos la información
```

## Valores únicos 

```sql
SELECT DISTINCT columna
FROM tabla;
```

## Contar elementos 

```sql
SELECT COUNT(columna)
FROM tabla;
```



## Ordenar elementos

```sql
SELECT *
FROM tabla
ORDER BY columna DESC;
```

## Groupby 

```sql
SELECT columna(s)
FROM tabla
GROUP BY column_name(s)
```

# Condiciones 

## Selección con condición

```sql
SELECT columna_1, columna_2
FROM nombre_tabla
WHERE columna_3='valor'; -- condición por la que queremos filtrar
```

## Selección por más de una condición 

`AND`

```sql
SELECT columna_1, columna_2
FROM nombre_tabla
WHERE columna_3='valor_1' AND columna_2='valor_2';
```

`OR`

```sql
SELECT columna_1, columna_2
FROM nombre_tabla
WHERE columna_3='valor_1' OR columna_2='valor_2';
```

`IN`

```sql
SELECT *
FROM nombre_tabla
WHERE columna_3 IN ('valor_1', 'valor_2', 'valor_3');
```


`NOT IN`

```sql
SELECT *
FROM nombre_tabla
WHERE columna_3 NOT IN ('valor_1', 'valor_2', 'valor_3');
```

`HAVING`

La cláusula HAVING se añadió a SQL porque la palabra clave WHERE no puede utilizarse con funciones de agregación.


```sql
SELECT columna(s)
FROM tabla
GROUP BY columna(s)
HAVING condición;
```

`BETWEEN`

El operador BETWEEN selecciona valores dentro de un rango determinado. Los valores pueden ser números, texto o fechas.

El operador BETWEEN es inclusivo: se incluyen los valores iniciales y finales.

```sql
SELECT columna(s)
FROM tabla
WHERE nombre_col BETWEEN valor1 AND valor2;
```

📌 **NOTA** En caso de que queramos poner un ORDER BY, GROUP BY y WHERE el orden será el siguiente:

```sql
SELECT columna(s)
FROM tabla
WHERE condicion
GROUP BY columna(s)
ORDER BY columna(s);
```

# Joins

## Inner Join 

La sentencia INNER JOIN selecciona los registros que tienen valores coincidentes en ambas tablas.




```sql
SELECT columna(s)
FROM tabla1
INNER JOIN tabla2
ON tabla1.nombre_col = tabla2.nombre_col;
```



## Left Join

La sentencia LEFT JOIN devuelve todos los registros de la tabla izquierda (tabla1), y los registros coincidentes de la tabla derecha (tabla2).



```sql
SELECT columna(s)
FROM tabla1
LEFT JOIN tabla2
ON tabla1.nombre_col = tabla2.nombre_col;
```

## Right Join

La sentencia RIGHT JOIN devuelve todos los registros de la tabla derecha (tabla2) y los registros coincidentes de la tabla izquierda (tabla1). 



```sql
SELECT columna(s)
FROM tabla1
RIGHT JOIN tabla2
ON tabla1.nombre_col = tabla2.nombre_col;
```

##  Union

LA sentencia UNION se utiliza para combinar el conjunto de resultados de dos o más sentencias SELECT.

- Cada sentencia SELECT dentro de UNION debe tener el mismo número de columnas


- Las columnas deben tener tipos de datos similares


- Las columnas de cada sentencia SELECT deben estar en el mismo orden


```sql
SELECT col1 FROM tabla1
UNION
SELECT col1 FROM tabla2
ORDER BY col1;

```
