## Estadística descriptiva



### Medidas de centralización

- `Media`: es el resultado de sumar todos los datos y dividirlo entre el número total de datos. Es una medida de tendencia central, es decir, el resultado del cálculo de una media dará un resultado ubicado en el centro de los valores con los que se ha calculado.


- `Mediana`: es el valor que ocupa el lugar central de una lista formada por todos los datos ordenados por su valor.



- `Moda`: es el número que más veces se repite dentro de nuestros datos.



- `Frecuencia`: la proporción de ocurrencia de un evento. Para eso usaremos un histograma


   
   

### Medidas de dispersión

- `Varianza`: es una medida de dispersión que representa la variabilidad de una serie de datos respecto a su media.


- `Desviación estándar`: la desviación estándar o desviación típica es una medida que ofrece información sobre la dispersión media de una variable. La desviación estándar es siempre mayor o igual que cero.




- `Rango`: es la diferencia entre el máximo y mínimo de la muestra. 


- `Cuartiles`: son cada uno de los tres valores que pueden dividir un grupo de números, ordenados de menor a mayor, en cuatro partes iguales. En otras palabras, cada cuartil determina la separación entre uno y otro subgrupo, dentro de un conjunto de valores estudiados. De esta forma podremos identificar 3 cuartiles:

    - Q1: representa el 25% de los datos

    - Q2: los que están por debajo de este cuartil representan el 50% de los datos.

    - Q3: representa el 75% de los datos.

# Distribuciones

Una distribución es una función que nos muestra la probabilidad de ocurrencia de un evento. Nos podemos encontrar con distintos tipos de distribuciones: 

- `Distribuciones discretas`: son aquellas que los eventos que intentamos describir son **variables discretas**, es decir, cara/cruz, los puntos de un dado, números de clientes que entran en nuestro local en un intervalo de tiempo determinado. 


    - Distribución de Bernoulli
    
    - Distribución binomial
    
    - Distribución Poisson
    

 

- `Distribuciones continuas`: son aquellas que los eventos que intentamos describir son **variables continuas**, es decir, peso de los alumnos, precio de las casas. 


    - Distribución normal
    
    - Distribución uniforme
    
    - Distribución exponencial

# Distribución Bernoulli

Tenemos dos posibles resultados, con una probabilidad de ocurrencia de 𝑝 y 1-𝑝. Es decir: 


Imaginemos que lanzamos una moneda, sus probabilidades serían:

- 𝑝: probabilidad de que salga cara


- 1-𝑝: probabilidad de que salga cruz

Configuramos nuestra distribución.

$$𝑋∼𝐵𝑒𝑟𝑛𝑜𝑢𝑙𝑙𝑖(𝑝)$$

```python
# probabilidad de sacar una manzana de una cesta de manzanas y naranjas

p_manzana = 0.8
print(f'la probabilidad de que salga una manzana es {p_manzana}')
# la probabilidad de sacar una naranja sería: 

p_naranja = 1 - 0.8
print(f'la probabilidad de que salga una naranja es {p_naranja}')
```

¿Cómo podemos generar este tipo de distribuciones en Python? Usando la librería `scipy.stats`

```python
# creamos una distribución para las manzanas

my_bernoulli = bernoulli(p_manzana)
```

```python
my_bernoulli
```

```python
# .rvs nos va a permitir generar números pseudoaleatorios que sigan la distribución que hemos creado 
muestra_10 = my_bernoulli.rvs(size=10)
muestra_10
```

```python
# y si aumentamos el tamaño de la muestra?

muestra_100 = my_bernoulli.rvs(size=100)
muestra_100
```

¿Cuántas veces ha salido la manzana? 

```python
muestra_100.sum()
```

```python
# como es un array podemos hacer una grafiquita para verlo mejor 😊

fig = sns.countplot(x = muestra_100, palette="Set3")
fig.set_xticklabels(['No manzana', 'Manzana']);
```

# Distribución Binomial


Describe el número de éxitos obtenidos al realizar $N$ experimentos con probabilidad de éxito $p$. 


Siguiendo el ejemplo de cara o cruz, queriendo que salga cara: 


- $p$: será probabilidad de que salga cara.

- $N$ número de experimentos, es decir, el número de veces que tiramos la moneda.


Se distribuye de la siguiente manera: 


   $X \sim Binomial(N, p)$


> La distribución de Bernoulli se puede expresar como una binomial de parámetro n = 1 y $p$

```python
# metemos 10 veces la mano en la bolsa
N = 10

# con una probabilidad de que salga una manzana de:
p_manzana = 0.8
```

```python
# iniciamos el método  para crear una binomial. En este caso le tendremos que pasar el numero de 
## experimentos(el número de veces que meteremos la mano en la cesta) y la probabilidad de sacar manzana

my_binomial = binom(N, p_manzana)
```

Llegados a este punto `scipy.stats` tiene una serie de métodos que nos van a permitir "jugar" con los parámetros de nuestras distribuciones. Esto son: 

- `.rvs`: número de repeticiones del experimento, es decir, cuantas veces que meto la mano en la bolsa de fruta. 


- `.pmf`: probabilidad de ocurrencia de un suceso dado del espacio muestral , es decir, la probabilidad de que salga manzana. 


- `.cdf`: probabilidad acumulada de un suceso


```python
# vamos a generar un array para simular el número de manzanas que queremos sacar:

numero_manzanas = np.arange(1,11)
```

```python
# hacemos un for para ver las distintas probabilidades en función del número de manzanas que podamos coger

for k in numero_manzanas:
    probab=binom.pmf(k=k,n=10,p=0.8)
    print(f"Probabilidad de que de mi bolsa de frutas con frutas aleatorias, {k} sean manzanas es: {probab}")
```

```python
# esto es lo mismo que el código anterior

probabilidades = my_binomial.pmf(numero_manzanas)
probabilidades
```

```python
# veamoslo graficamente

plt.plot(numero_manzanas, my_binomial.pmf(numero_manzanas), 'o')
plt.vlines(numero_manzanas, 0, probabilidades, colors='black', lw=5, alpha=0.5);
```

```python
# y que pasa con el acumulado? Recordamos, nos devuelve la probabilidad acumulada de que ocurra un suceso

for k in numero_manzanas:
    acumulada = binom.cdf(k = k, n= N, p = p_manzana)
    print(f"La probabilidad acumulada de sacar {k} manzanas es: {acumulada}")
```

De nuevo, este código es lo mismo que lo anterior de una forma más sencilla 👇🏽


```python
probabilidades_acumuladas = my_binomial.cdf(numero_manzanas)
probabilidades_acumuladas
```

```python
# visualmente: 

plt.plot(numero_manzanas, my_binomial.cdf(numero_manzanas), 's')
plt.vlines(numero_manzanas, 0, probabilidades_acumuladas, colors='black', lw=5, alpha=0.5);
```

¿Pero en realidad esto que significa?

```python
my_binomial.cdf(3)
```

🚨 Significa que la probabilidad de coger al menos tres manzanas es de 0.0008. O lo que es lo mismo, la probabilidad de coger más de 3 manzanas es 0.999

# Distribución Poisson

La distribución de Poisson se utiliza para describir el número de veces que puede ocurrir un evento **en un intervalo de tiempo concreto**.

Ejemplos:

 * El número de meteoritos de más de 1 metro de diámetro que chocan contra la Tierra en un año.
 * El número de pacientes que llegan a una sala de urgencias entre las 10 y las 11 de la noche

 
Se distribuye de la siguiente manera: 

$X \sim Po(\lambda)$

Donde: 

- $\lambda$: es la media del número de ocurrencias, es decir, en media entran X pacientes a la hora en el hospital. 

```python
# definimos la lambda. De media entran 6 pacientes en el hospital

pacientes = 6
```

```python
# iniciamos la distribución de Poisson

my_poisson = poisson(mu=pacientes)
```

```python
# generamos 30 eventos pseudoaleatorios

my_poisson.rvs(size=30)
```

```python
# lo primero, generamos un array con números aleatorios de pacientes que entran en el hospital

pacientes_entran = np.arange(0,15)
pacientes_entran
```

```python
probabilidades_binomial = my_poisson.pmf(pacientes_entran)
probabilidades_binomial
```

```python
# veamoslo visualmente

plt.plot(pacientes_entran, my_poisson.pmf(pacientes_entran), 'bo')
plt.vlines(pacientes_entran, 0, probabilidades_binomial, colors='b', lw=5, alpha=0.5);
```

¿Y qué pasa con el acumulado? 

```python
my_poisson.cdf(5)
```

La probabilidad de que entren menos de 5 pacientes es de 0.44. 
Por el contrario, la probilidad de que entren más de 0.44 es de 0.55
